```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show

```

```python
#Reading the parquet file using pandas
trips = pd.read_parquet('C:/Users/Arpita/Downloads/yellow_tripdata_2022-01.parquet',engine="auto",columns=None)
```

```python
from datetime import datetime
trips['Weekday'] = trips["tpep_pickup_datetime"].dt.strftime('%A')
trips.head()
```

```python
pay1={"Credit Card":0,"Cash":0,"No Change":0,"Dispute":0,"Unknown":0}
paymenttypes = {1:'Credit Card', 2:'Cash', 3:'No change', 4:'Dispute', 5:'Unknown'}
trips['payment_types_explicit']= trips['payment_type'].map(paymenttypes)
print(pay1.keys())
```

```python
#pay=dict(df["payment_types_explicit"].value_counts())
busydays = dict(trips.groupby(by="Weekday")["payment_types_explicit"].value_counts())
print(busydays)
```

```python
#Storing the frequency of usage of payment methods by weekdays in lists 
monday=[291039,79212,1917,1650,0]
tuesday=[245408,68044,1605,1351,0]
wednesday=[257530,69884,1577,1445,0]
thursday=[272907,70512,1669,1477,0]
friday=[278008,69708,1692,1483,0]
saturday=[278113,70736,1671,1697,0]
sunday=[261869,66875,1578,1570,1]
```

```python
from bokeh.models import ColumnDataSource
from bokeh.palettes import GnBu7
from bokeh.plotting import figure, show

paym = list(pay1.keys())
days = ["monday", "tuesday", "wednesday", "thursday", "friday", "saturday", "sunday"]

weeklist = {'payment_methods' : paym,
           "monday"   : monday,
           "tuesday"   : tuesday,
           "wednesday" : wednesday,
           "thursday" : thursday,
            "friday"  : friday,
            "saturday": saturday,
            "sunday" : sunday}


p = figure(y_range=paym, height=350, x_range=(0,2000000),x_axis_label="Frequency of usage",y_axis_label="payment method", title="Payment methods by days of week",
           toolbar_location=None)

p.hbar_stack(days, y='payment_methods', height=0.9, color=GnBu7, source=ColumnDataSource(weeklist),
             legend_label=[f"{day}" for day in days])

p.y_range.range_padding = 0.1
p.ygrid.grid_line_color = None
p.legend.location = "top_right"
p.axis.minor_tick_line_color = None
p.outline_line_color = None

show(p)
```
