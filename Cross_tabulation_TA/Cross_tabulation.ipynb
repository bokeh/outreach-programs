```python
import pandas as pd
```

```python
#Reading the parquet file using pandas
trips = pd.read_parquet('C:/Users/Arpita/Downloads/yellow_tripdata_2022-01.parquet',engine="auto",columns=None)
#Taking the subset of first 70 rowws of the given data
df=trips.head(100)
#print(df)
```

```python
#checking for the enries with null values
print(trips.info(null_counts=True))
#removing the null values
trips.dropna(how="any",inplace = True)
print(trips.info(null_counts=True))
```

```python
from datetime import datetime
trips['Weekday'] = trips["tpep_pickup_datetime"].dt.strftime('%A')
trips.head()
```

```python

pay1={"Credit Card":0,"Cash":0,"No Change":0,"Dispute":0,"Unknown":0}
paymenttypes = {1:'Credit Card', 2:'Cash', 3:'No change', 4:'Dispute', 5:'Unknown'}
trips['payment_types_explicit']= trips['payment_type'].map(paymenttypes)
```

```python
df=trips
```

```python
from bokeh.core.properties import value
from bokeh.plotting import ColumnDataSource, figure, show
from bokeh.transform import cumsum, factor_cmap

rows = pd.crosstab(df.Weekday, df.payment_types_explicit, aggfunc='sum', values=df.passenger_count, normalize="index")

source = ColumnDataSource(rows.T)

weekd = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
pte = source.data["payment_types_explicit"]

p = figure(y_range=weekd, x_range=(-0.55, 1.2), height=600, width=1000, tools="",
           x_axis_location=None, toolbar_location=None, outline_line_color=None)
p.grid.grid_line_color = None
p.yaxis.fixed_location = 0
p.axis.major_tick_line_color = None
p.axis.major_label_text_color = None
p.axis.axis_line_color = "#4a4a4a"
p.axis.axis_line_width = 6

source.data["color"] = [ "#dadada","#dadada", "#4a4a4a", "#dadada","#4a4a4a","#dadada"]
for y in weekd:
    left, right = cumsum(y, include_zero=True), cumsum(y)

    p.hbar(y=value(y), left=left, right=right, source=source, height=0.9,
           color=factor_cmap("payment_types_explicit", "MediumContrast4", pte))

    pcts = source.data[y]
    source.data[f"{y} text"] = [f"{r}\n{x*100:0.1f}%" for r, x in zip(pte, pcts)]

    p.text(y=value(y), x=left, text=f"{y} text", source=source, x_offset=10,
           text_color="color", text_baseline="middle", text_font_size="12px")

totals = pd.crosstab(df.Weekday, df.payment_types_explicit, margins=True, aggfunc='sum',
                     values=df.passenger_count, normalize="columns").All

p.hbar(right=0, left=-totals, y=totals.index, height=0.9, color="#dadada")

text = [f"{name} ({totals.loc[name]*100:0.1f}%)" for name in weekd]
p.text(y=weekd, x=0, text=text, text_baseline="middle", text_align="right",
       x_offset=-12, text_color="#4a4a4a", text_font_size="15px",
       text_font_style="bold")

show(p)
```
